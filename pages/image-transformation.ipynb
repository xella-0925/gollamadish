```python
import numpy as np #Import libraries needed
import cv2
import matplotlib.pyplot as plt
```

```python
img_ = cv2.imread("jeno-lee.jpg") #use imread to load an image from the specified file and return it
img_ = cv2.cvtColor(img_, cv2.COLOR_BGR2RGB) #use cvtColor to convert image from one color space to another.
                                             # cv2. COLOR_BGR2RGB returns image in RGB format.
```

```python
rows, cols, dims = img_.shape #define img.shape 
                              #img.shape returns a tuple of the number of rows, columns
```

```python
def plt_grph(transformed_img_): #create a function to plot transformed image image
    plt.axis('off') #hide axes and borders
    plt.imshow(transformed_img_) #display data as an image
```

```python
def translate (img_, rows, cols): #create def function to translate image and define its parameters.
    m_translation_ = np.float32 ([[1, 0, 100], #create an array to determine the number of rows and columns of an image.
                                  [0, 1, 50],
                                  [0, 0, 1]])
                                  
    translated_img_ = cv2.warpPerspective(img_, m_translation_, (cols, rows)) #return an image whose size is 
                                                                              #the same as the size of the original image. 

    return translated_img_ 
```

```python
result = translate (img_, rows, cols) # create a new variable called "result" and assign the function "translate" to it
                                      #so that the function can easily be called, which inreases code reusability. 
plt_grph(result) #call graph function and add "result" as its parameter to dispaly image.
```

```python
def plt_grph(scaled_img_): #create a function to plot scaled image 
    plt.axis('off')   #hide axes and borders
    plt.imshow(scaled_img_) #display data as an image
```

```python
def scaling (img_, rows, cols): #create def function to scale image and define its parameters.
    m_scaling_ = np.float32 ([[1.5, 0, 100], #create an array to determine the number of rows and columns of an image.
                              [0, 1.8, 0],
                              [0, 0, 1]])
                                  
    scaled_img_ = cv2.warpPerspective(img_, m_scaling_, (cols*2, rows*2)) # multiply cols and rows to get desired size
                                                                          #returns an image whose size is 
                                                                          #the same as the size of the original image.

    return scaled_img_
```

```python
result = scaling (img_, rows, cols) #create a new variable called "result" and assign the function "scaling" to it
                                    #so that the function can easily be called, which inreases code reusability
plt_grph(result) #call graph function and add "result" as its parameter to dispaly image.
```

```python
def plt_grph(rotated_img_): #create a function to plot rotated image 
    plt.axis('off') #hide axes and borders
    plt.imshow(rotated_img_) #display data as an image
```

```python
def rotate (img_, rows, cols): #create def function to rotate image and define its parameters.
    angle = np.radians(10)
    
    m_rotation_ = np.float32 ([[np.cos(angle), (-np.sin(angle)), 0], #create an array to determine the 
                          [np.sin(angle), np.cos(angle), 0],         #number of rows and columns of an image.
                          [0, 0, 1]])

    rotated_img_ = cv2.warpPerspective(img_, m_rotation_, (int(cols), int(rows))) #use warPerspective to apply perspective 
                                                                                  #tranformation the image. 
    m_rotation_ = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1) # rotate image by 90 degrees with respect to center without 
                                                        # any scaling. use cv2.getRotationMatrix2D() to get the rotation matrix 
    rotated_img_ = cv2.warpAffine(img_, m_rotation_, (cols, rows)) #warpAffine applies the affine transformation 

    return rotated_img_ #display result
```

```python
result = rotate (img_, rows, cols) #create a new variable called "result" and assign the function "rotate" to it
plt_grph(result) #call graph function and add "result" as its parameter to dispaly image.
```

```python
def plt_grph(reflected_img_): #create a function to plot reflected image 
    plt.axis('off') #hide axes and borders
    plt.imshow(reflected_img_) #display data as an image
```

```python
def flip (img_, rows, cols): #create def function to flip image and define its parameters.               
    m_reflection_ = np.float32 ([[1, 0, 0  ], #create an array to determine the 
                             [0, -1, rows],   #number of rows and columns of an image.
                             [0, 0, 1  ]])

    reflected_img_ = cv2.warpPerspective(img_, m_reflection_, (int(cols), int(rows)))  #use warPerspective to apply perspective 
                                                                                       #tranformation the image. 
    img_flipped_ = cv2.flip(img_, 1) # use cv2.flip to flip image (1 means flipping around y-axis)

    return reflected_img_ #display result 
```

```python
result = flip (img_, rows, cols) #create a new variable called "result" and assign the function "flip" to it
plt_grph(result) #call graph function and add "result" as its parameter to dispaly image.
```

```python
def plt_grph(sheared_img_x): #create a function to plot sheared image x 
    plt.axis('off') #hide axes and borders
    plt.imshow(sheared_img_x) #display data as an image
```

```python
def shear (img_, rows, cols): #create def function to shear image (x) and define its parameters.                           
    m_shearing_x = np.float32 ([[1, 0.5, 0], #create an array to determine the 
                           [0, 1  , 0],      #number of rows and columns of an image.
                           [0, 0  , 1]])     #shearing applied to x-axis

    sheared_img_x = cv2.warpPerspective(img_, m_shearing_x, (int(cols*1.5), int(rows*1.5))) #use warPerspective to apply perspective 
                                                                                            #transformation to the image.
    return sheared_img_x #display result 
```

```python
result = shear (img_, rows, cols) #create a new variable called "result" and assign the function "shear" to it
plt_grph(result) #call graph function and add "result" as its parameter to dispaly image.
```
